```python
#PEFeatureExtractor è la classe di Ember che estrae caratteristiche dai file PE per machine learning.
from ember.features import PEFeatureExtractor
import pandas as pd

#Extract a feature vector from a binary file using the ember Feature Extractor
def get_ember_features(path, decompress = False):
    extractor = PEFeatureExtractor(print_feature_warning=False) #Crea un oggetto estrattore.
    if decompress:
      bytes = decompress_file(path)
    else:
      with open(path, 'rb') as f:
        bytes = f.read()

    features = [float(value) for value in extractor.feature_vector(bytes)] #Applica l’estrattore Ember al file binario e converte tutti i valori in float.
    return features


import zlib
def decompress_file(input_file_path):
    # read the file
    with open(input_file_path, 'rb') as f:
        compressed_data = f.read()

    # decompress the file with zlib
    decompressed_data = zlib.decompress(compressed_data)

    return decompressed_data
```

# Estrazione features files training set

```python

# Inizializza due liste vuote per salvare le feature dei file benigni e quelle dei file malware.
benign_features = []
malware_features = []

benign_dir = '/content/drive/Shareddrives/AI for Cybersecurity/Consegna/Dataset1/training/benign'
malware_dir = '/content/drive/Shareddrives/AI for Cybersecurity/Consegna/Dataset1/training/malware'
bad_files = []


#Per ogni file nella cartella dei benigni:
#os.listdir(...) prende tutti i nomi dei file.
#os.path.join(...) costruisce il percorso completo del file.
#get_ember_features(...) estrae il vettore di feature Ember da quel file (usando il PEFeatureExtractor).
#Lo aggiunge alla lista benign_features.
#import pefile
#import os


#for filename in os.listdir(benign_dir):
#    path = os.path.join(benign_dir, filename)
#    try:
#        pe = pefile.PE(path)
#        print(f"[OK] Riconosciuto come PE: {filename}")
#    except Exception as e:
#        print(f"[NO] {filename} non è un PE valido: {e}")
skipped = 0
parsed = 0

# Estrazione dai file benigni
print("🔍 Estrazione feature dai file benigni...")
for filename in os.listdir(benign_dir):
    path = os.path.join(benign_dir, filename)
    try:
        feats = get_ember_features(path)
        if feats:
            benign_features.append(feats)
            parsed += 1
    except Exception as e:
        bad_files.append((filename, str(e)))
        skipped += 1

# Estrazione dai file malware
print("🔍 Estrazione feature dai file malware...")
for filename in os.listdir(malware_dir):
    path = os.path.join(malware_dir, filename)
    try:
        feats = get_ember_features(path)
        if feats:
            malware_features.append(feats)
            parsed += 1
    except Exception as e:
        bad_files.append((filename, str(e)))
        skipped += 1

# Salvataggio
with open("/content/drive/Shareddrives/AI for Cybersecurity/Consegna/Dataset1/training/benign_features.pkl", "wb") as f:
    pickle.dump({
        "features": benign_features,
        "label": 0,
        "count": len(benign_features)
    }, f)
print(f"✅ Salvate {len(benign_features)} feature da file benigni in training/benign_features.pkl")

with open("/content/drive/Shareddrives/AI for Cybersecurity/Consegna/Dataset1/training/malware_features.pkl", "wb") as f:
    pickle.dump({
        "features": malware_features,
        "label": 1,
        "count": len(malware_features)
    }, f)
print(f"✅ Salvate {len(malware_features)} feature da file malware in training/malware_features.pkl")

with open("/content/drive/Shareddrives/AI for Cybersecurity/Consegna/Dataset1/training/bad_files.pkl", "wb") as f:
    pickle.dump(bad_files, f)
print(f"⚠️ Salvati {len(bad_files)} file problematici in training/bad_files.pkl")

print(f"\n🟢 Totale parsati: {parsed} | 🔴 Skippati: {skipped}")


#Quelli che ricevi (es. Unable to find the section associated with BASE_RELOCATION_TABLE, Relocation corrupted, Not a valid DLL name) sono warning comuni quando si analizzano
# file PE corrotti, obfuscati o non standard, cosa normale nei malware.
```
